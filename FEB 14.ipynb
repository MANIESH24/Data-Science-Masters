```python
##4)
import threading

def print_squares():
    for i in range(1, 6):
        square = i ** 2
        print("Square:", square)

def print_cubes():
    for i in range(1, 6):
        cube = i ** 3
        print("Cube:", cube)

# Create thread one
thread1 = threading.Thread(target=print_squares)

# Create thread two
thread2 = threading.Thread(target=print_cubes)

# Start both threads
thread1.start()
thread2.start()

# Wait for both threads to finish
thread1.join()
thread2.join()

print("Program completed.")
```
